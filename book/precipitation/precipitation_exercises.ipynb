```{exercise-start}
:label: ex_prec_7
```
Compute the semivariance between two raingauges and use {numref}`fig:PRE_modelling_sphericalvariogram` to estimate the distance between the two stations.

| $R_1 \: [\text{mm}]$ | $R_2 \: [\text{mm}]$ |
| --- | --- |
| 10.0 | 12.2 |
| 3.8 | 5.1 |
| 6.5 | 2.2 |
| 0 | 0.1 |
| 3.8 | 1.4 |
| 12.0 | 16.5 |

**Remember that**: $Var(X) = \frac{1}{n-1} \cdot \sum_{i=1}^{n} (x_{i}-\mu)^2$.

```{exercise-end}
```

:::{dropdown} Answer&nbsp;{ref}`ex_prec_7`
Compute the difference $R_1 - R_2$:

| $R_1$ | $R_2$ | $R_1 - R_2$ |
| --- | --- | --- |
| 10.0 | 12.2 | -2.2 |
| 3.8 | 5.1 | -1.3 |
| 6.5 | 2.2 | 4.3 |
| 0 | 0.1 | -0.1 |
| 3.8 | 1.4 | 2.4 |
| 12.0 | 16.5 | -4.5 |

The mean of $R_1 - R_2$ is:

$$
\mu = \frac{-2.2-1.3+4.3-0.1+2.4-4.5}{6} = -0.23
$$

The variance of $R_1 - R_2$ is:

$$
Var(X) = \frac{1}{n-1} \cdot \sum_{i=1}^{n} (x_{i}-\mu)^2
$$
$$
Var(X) = \frac{1}{5} \cdot [(-2.2--0.23)^2 + (-1.3--0.23)^2 + (4.3--0.23)^2 + (-0.1--0.23)^2 \\ + (2.4--0.23)^2 + (-4.5--0.23)^2] = 10.1 \: \text{mm}^2
$$

Using {eq}`semivariogram`:

$$
\gamma(d) = \frac{1}{2} Var[R(p_1)-R(p_2)] = \frac{1}{2} \cdot 10.1 = 5.1 \: \text{mm}^2
$$

Read from {numref}`fig:PRE_modelling_sphericalvariogram`: $5 \: \text{mm}^2$ corresponds to about 10 km. Of course this type of statistics should be performed with much larger datasets.
:::

```{exercise-start}
:label: ex_prec_9
```
Rainfall can be measured by means of (weather) radar.

a) The so-called Marshall-Palmer relation ($Z=200R^{1.6}$) is often used to relate radar reflectivity $Z$ [$\frac{\text{mm}^6}{\text{m}^{3}}$] and rainfall intensity $R$ [$\frac{\text{mm}}{\text{h}}$]. Suppose a weather radar measures a reflectivity of 1000 $\frac{\text{mm}^6}{\text{m}^{3}}$. What is the corresponding rainfall rate? And what is the rainfall rate for $Z$=100,000 $\frac{\text{mm}^6}{\text{m}^{3}}$?

b) Since $Z$ can vary strongly over orders of magnitude while this is not the case for $R$, it is common to express $Z$ in decibels (dBZ). This is a logarithmic scale defined as $10 \times \log(Z)$, with $\log$ the logarithm with base 10. Suppose it rains with an intensity of 10 $\frac{\text{mm}}{\text{h}}$, what reflectivity (in dBZ) will be measured by the weather radar?
```{exercise-end}
```



:::{dropdown} Answer&nbsp;{ref}`ex_prec_9`a
$$
R = (Z / 200) ^{1 / 1.6} = 5 ^{0.625} = 2.7 \text{ mm/h}
$$

$$
R = 500 ^{0.625} = 48.6 \text{ mm/h}
$$
:::

:::{dropdown} Answer&nbsp;{ref}`ex_prec_9`b
$Z = 200 R ^{1.6}$, so $10\log{Z} = 10\log{200} + 16\log{R} = 23 + 16 = 39 \text{ dBZ}$
:::
